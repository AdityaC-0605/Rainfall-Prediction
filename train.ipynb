```python
# Melt the dataframe to convert month columns into rows
dates = df.melt(id_vars=['YEAR'], 
                var_name='month', 
                value_name='temp')

# Create proper datetime column
dates['date'] = pd.to_datetime(dates['month'] + ' ' + dates['YEAR'].astype(str), format='%b %Y')

# Extract year and month name
dates['year'] = dates['date'].dt.year
dates['month_name'] = dates['date'].dt.month_name()

# Sort by date and reset index
dates = dates.sort_values('date').reset_index(drop=True)

dates.head()
```

```python
dates['season'] = np.empty(shape= dates.shape[0])

i = 0
for m in dates['month']:
    season = None
    if m in ['November', 'December', 'January', 'February']:
        season = 'winter'
    elif m in ['March', 'April', 'May']:
        season = 'summer'
    elif m in ['June', 'July', 'August', 'September', 'October']:
        season = 'monsoon'
    
    dates['season'][i] = season
    i += 1

dates.head()
```

```python
sns.relplot(x= dates['date'], y= dates['temp'], hue= dates['month'], legend= True)
plt.show()
```
